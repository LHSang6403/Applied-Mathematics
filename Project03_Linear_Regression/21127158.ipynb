```python
import pandas as pd
import numpy as np

# Import thêm dữ thư viện nếu cần
from sklearn.model_selection import KFold
```

# Đọc dữ liệu

```python
# Đọc dữ liệu bằng pandas
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Lấy các đặc trưng X và giá trị mục tiêu y cho các tập huấn luyện (train) và kiểm tra (test)
X_train = train.iloc[:, :-1]    # Dataframe (chứa các đặc trưng huấn luyện)
y_train = train.iloc[:, -1]     # Series    (chứa 1 giá trị mục tiêu kiểm tra)

X_test = test.iloc[:, :-1]      # Dataframe (chứa các đặc trưng kiểm tra)
y_test = test.iloc[:, -1]       # Series    (chứa 1 giá trị mục tiêu kiểm tra)

# print('X\n', X_test)
# print('y\n', y_test)

# Sinh viên có thể sử dụng các khác nếu cần
```

# Cài đặt hàm

```python
# Cài đặt các hàm cần thiết ở đây
class OLSLinearRegression:
    def fit(self, X, y):
        X_pinv = np.linalg.inv(X.T @ X) @ X.T    # np.linalg.pinv(X)
        self.w = X_pinv @ y

        return self

    def get_params(self):
        return self.w

    def predict(self, X):
        return np.sum(self.w.ravel() * X, axis=1) 

def model_mae(y_test, y_pred):
    if len(y_test) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    
    absolute_errors = [abs(actual - predicted) for actual, predicted in zip(y_test, y_pred)]
    mae = sum(absolute_errors) / len(y_test)
    return mae
```

# Yêu cầu 1a: Sử dụng toàn bộ 11 đặc trưng đầu tiên `Gender`, `10percentage`, `12percentage`, `CollegeTier`, `Degree`, `collegeGPA`, `CollegeCityTier`, `English`, `Logical`, `Quant`, `Domain` (2 điểm) 

```python
# Phần code cho yêu cầu 1a
X_train1a = X_train[[
    'Gender',
    '10percentage',
    '12percentage',
    'CollegeTier',
    'Degree', 
    'collegeGPA',
    'CollegeCityTier',
    'English',
    'Logical',
    'Quant',
    'Domain',
]]

X_test1a = X_test[[
    'Gender', 
    '10percentage', 
    '12percentage', 
    'CollegeTier', 
    'Degree', 
    'collegeGPA',
    'CollegeCityTier',
    'English',
    'Logical',
    'Quant',
    'Domain',
]]

# To train 11 fearures
def train_1a(X_train, y_train, X_test):
    w = OLSLinearRegression().fit(X_train, y_train)
    y_pred = w.predict(X_test)
    return y_pred, w

def question_1a_MAE_check(y_test, y_pred, w):
    return model_mae(y_test, y_pred), np.array(w.get_params()).reshape(-1, 1)
```

```python
# Gọi hàm MAE (tự cài đặt hoặc từ thư viện) trên tập kiểm tra
y_pred, w = train_1a(X_train1a.to_numpy(), y_train.to_numpy(), X_test1a.to_numpy())
mae, w = question_1a_MAE_check(y_test.to_numpy(), y_pred, w)

print("Weight:\n", np.round(w, 3), "\nMAE:", np.round(mae, 3))
```

# Yêu cầu 1b: Xây dựng mô hình sử dụng duy nhất 1 đặc trưng tính cách với các đặc trưng tính cách gồm `conscientiousness`, `agreeableness`, `extraversion`, `nueroticism`, `openess_to_experience`, tìm mô hình cho kết quả tốt nhất (1 điểm)

Lưu ý: khi sử dụng cross-validation, sinh viên cần xáo trộn dữ liệu 1 lần duy nhất và thực hiện trên toàn bộ đặc trưng

```python
# Phần code cho yêu cầu 1b
# Tìm ra đặc trưng tốt nhất
# In ra các kết quả cross-validation như yêu cầu

features_1b = [
    'conscientiousness',
    'agreeableness',
    'extraversion',
    'nueroticism',
    'openess_to_experience',
    'Salary'
]

def question_1b_handler(train, features, k_fold=5):
    mae_arr = np.zeros(train.shape[1] - 1)
    shuffle_data = KFold(n_splits = k_fold, shuffle = True)

    for train_split, test_split in shuffle_data.split(train):
        i = 0
        for column in features:
            if column != 'Salary':
                feature_train = train.iloc[train_split].loc[:,[column]]
                label_train = train.iloc[train_split].loc[:,['Salary']]

                feature_test = train.iloc[test_split].loc[:,[column]]
                label_test = train.iloc[test_split].loc[:,['Salary']]
    
                y_pred, w = train_1a(feature_train.to_numpy(), label_train.to_numpy(), feature_test.to_numpy())
                mae, w = question_1a_MAE_check(label_test.to_numpy(), y_pred, w)

                mae_arr[i] += mae
                i += 1

    mae_arr /= k_fold
    print('MAE: ', np.round(mae_arr, 3))

    return np.round(mae_arr, 3), features[np.argmin(mae_arr)]
```

```python
train_1b = train[features_1b]
mae_arr_1b, best_feature_1b = question_1b_handler(train_1b, features_1b, 5)

print("Best feature: ", best_feature_1b)
```

```python
# Huấn luyện lại mô hình best_personality_feature_model với đặc trưng tốt nhất trên toàn bộ tập huấn luyện
features_train_1b = train.loc[:, [best_feature_1b]]
features_test_1b = test.loc[:, [best_feature_1b]]

y_pred_1b, w_1b = train_1a(features_train_1b.to_numpy(), y_train.to_numpy(), features_test_1b.to_numpy())
```

```python
# Gọi hàm MAE (tự cài đặt hoặc từ thư viện) trên tập kiểm tra với mô hình best_personality_feature_model
mae_1b, w_1b = question_1a_MAE_check(y_test.to_numpy(), y_pred_1b, w_1b)

print("Weight:\n", np.round(w_1b, 3), "\nMAE:", np.round(mae_1b, 3))
```

Công thức hồi quy (phần trọng số làm tròn đến 3 chữ số thập phân, ví dụ 0.012345 $\to$ 0.012)

# Yêu cầu 1d: Sinh viên tự xây dựng mô hình, tìm mô hình cho kết quả tốt nhất (3 điểm)

Lưu ý: khi sử dụng cross-validation, sinh viên cần xáo trộn dữ liệu 1 lần duy nhất và thực hiện trên toàn bộ $m$ mô hình mà sinh viên thiết kế

## Tìm mô hình

```python
# Trình bày các phần tìm ra mô hình
```

```python
def data_constructor(data, option=0):
    if(option == 0): # based on grades at university
        return data.loc[:, [
            "collegeGPA",
            "ComputerScience",
            "ComputerProgramming",
            "ElectronicsAndSemicon",
            "TelecomEngg",
            "Salary",
        ]]
    elif(option == 1): # based on all AMCAT scores and sqrt of numbers directly related to professional skills
        return pd.DataFrame([[np.sqrt(np.abs(d[i])) if i == 3 or i == 4 or i == 5 or i == 6 or i == 7 or i == 8 or i == 9 or i == 10
                               else d[i] for i in range(17)] for d in ((data.loc[:,[
            "English",
            "Logical",
            "Quant",
            "Domain", # sqrt
            "ComputerProgramming", # sqrt
            "ElectronicsAndSemicon", # sqrt
            "ComputerScience", # sqrt
            "MechanicalEngg", # sqrt
            "ElectricalEngg", # sqrt
            "TelecomEngg", # sqrt
            "CivilEngg", # sqrt
            "conscientiousness",
            "agreeableness",
            "extraversion",
            "nueroticism",
            "openess_to_experience",
            "Salary",
        ]
        ]).to_numpy())], columns = [
            "English",
            "Logical",
            "Quant",
            "Domain",
            "ComputerProgramming",
            "ElectronicsAndSemicon",
            "ComputerScience",
            "MechanicalEngg",
            "ElectricalEngg",
            "TelecomEngg",
            "CivilEngg",
            "conscientiousness",
            "agreeableness",
            "extraversion",
            "nueroticism",
            "openess_to_experience",
            "Salary",
        ])
    elif(option == 2): # based on all AMCAT scores and squares of numbers directly related to professional skills
        return pd.DataFrame([[d[i] ** 2 if i == 1 or i == 3 or i == 4 or i ==5 or i == 6 or i == 7 or i == 8 or i == 9 or i == 10 or i == 13
                               else d[i] for i in range(17)] for d in ((data.loc[:,[
            "English",
            "Logical", # ** 2
            "Quant",
            "Domain", # ** 2
            "ComputerProgramming", # ** 2
            "ElectronicsAndSemicon", # ** 2
            "ComputerScience", # ** 2
            "MechanicalEngg", # ** 2
            "ElectricalEngg", # ** 2
            "TelecomEngg", # ** 2
            "CivilEngg", # ** 2
            "conscientiousness",
            "agreeableness",
            "extraversion", # ** 2
            "nueroticism",
            "openess_to_experience",
            "Salary",
        ]
        ]).to_numpy())], columns = [
            "English",
            "Logical",
            "Quant",
            "Domain",
            "ComputerProgramming",
            "ElectronicsAndSemicon",
            "ComputerScience",
            "MechanicalEngg",
            "ElectricalEngg",
            "TelecomEngg",
            "CivilEngg",
            "conscientiousness",
            "agreeableness",
            "extraversion",
            "nueroticism",
            "openess_to_experience",
            "Salary",
        ])
    elif(option == 3): # based on AMCAT major scores and squares of scores of the first 4 parameters
        return pd.DataFrame([[d[i] ** 2 if i == 0 or i == 1 or i == 2 or i == 3
                               else d[i] for i in range(12)] for d in ((data.loc[:,[
            "English", # ** 2
            "Logical", # ** 2
            "Quant", # ** 2
            "Domain", # ** 2
            "ComputerProgramming", 
            "ElectronicsAndSemicon",
            "ComputerScience",
            "MechanicalEngg",
            "ElectricalEngg",
            "TelecomEngg",
            "CivilEngg",
            "Salary",
        ]
        ]).to_numpy())], columns = [
            "English",
            "Logical",
            "Quant",
            "Domain",
            "ComputerProgramming",
            "ElectronicsAndSemicon",
            "ComputerScience",
            "MechanicalEngg",
            "ElectricalEngg",
            "TelecomEngg",
            "CivilEngg",
            "Salary",
        ])
    elif(option == 4): # based on scores of the previous exams
        return data.loc[:, [
            "10percentage",
            "12percentage",
            "collegeGPA",
            "Salary",
        ]]
    elif(option == 5): # based on only collegeGPA
        return data.loc[:, [
            "collegeGPA",
            "Salary",
        ]]
    else:
        print('Error option')
        return data
    
def question_1d_handler(train, k_fold = 5, number_of_models = 5):
    mae_arr = np.zeros((number_of_models, 1))
    shuffle_data =  KFold(n_splits=k_fold, shuffle=True)

    for train_split, test_split in shuffle_data.split(train):
        # print("Train split", train_split, "Test split", test_split)
        for i in range(number_of_models):

            model_data = data_constructor(train, i)
            # print("Loop", i, model_data)

            features_train = model_data.iloc[train_split].iloc[:, :-1]
            labels_train = model_data.iloc[train_split].loc[:,['Salary']]

            features_test = model_data.iloc[test_split].iloc[:, :-1]
            labels_test = model_data.iloc[test_split].loc[:,['Salary']]
                            
            y_pred, w = train_1a(features_train.to_numpy(), labels_train.to_numpy(), features_test.to_numpy())
            mae, w = question_1a_MAE_check(labels_test.to_numpy(), y_pred, w)

            mae_arr[i] += mae
            i += 1
            
    mae_arr /= k_fold

    print('MAE:\n', np.round(mae_arr, 3))
    return mae_arr, np.argmin(mae_arr)
```

## Thử nghiệm, so sánh các mô hình

```python
# Phần code cho yêu cầu 1d
# Tìm ra mô hình tốt nhất (tự thiết kế bởi sinh viên)
# In ra các kết quả cross-validation như yêu cầu

m = 6 # models

mae_arr_1d, best_1d = question_1d_handler(train, k_fold = 5, number_of_models = m)

print("My best model: number", best_1d + 1)
```
